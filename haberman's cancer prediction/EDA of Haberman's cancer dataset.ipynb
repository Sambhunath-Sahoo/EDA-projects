### OBJECTIVE  : predict whether the patent will survive by the given treatment or not

# visualization 

```python
sns.FacetGrid(df, hue="Survival_Status", size=5)\
.map(sns.distplot, "Operation_Year").add_legend();
plt.show();
```

```python
sns.FacetGrid(df, hue="Survival_Status", size=5)\
.map(sns.distplot, "Age_Of_Patient").add_legend();
plt.show();
```

```python
sns.FacetGrid(df, hue="Survival_Status", size=5)\
.map(sns.distplot, "Axil_nodes").add_legend();
plt.show();
```

```python
sns.set_style("whitegrid")
sns.pairplot(df, hue="Survival_Status", vars=['Age_Of_Patient','Operation_Year','Axil_nodes'],size=3)
```

### Observation :

  The data are highly mixed up, none of the variable-pairs can help us find linearly separable clusters hence we can't find "lines" and "if-else" conditions to build a simple model to classify the survive status of the patient.
